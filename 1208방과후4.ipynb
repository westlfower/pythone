```python
import pandas as pd
import numpy as np

np.random.seed(42)
names1 = ['민준', '서연', '도윤', '하윤', '지호', '예지', '준혁', '수민', '현우', '지은']
names2 = ['유준', '지유', '시우', '서윤', '민준', '하린', '지후', '채원', '도윤', '소율']

data1 = {
    '번호': range(1, 11),
    '반': ['1반']*10,
    '이름': names1,
    '성별': ['남', '여', '남', '여', '남', '여', '남', '여', '남', '여'],
    '수학': np.random.randint(60, 101, 10),
    '영어': np.random.randint(60, 101, 10),
    '과학': np.random.randint(60, 101, 10)
}
data2 = {
    '번호': range(1, 11),
    '반': ['2반']*10,
    '이름': names2,
    '성별': ['남', '여', '남', '여', '남', '여', '남', '여', '남', '여'],
    '수학': np.random.randint(60, 101, 10),
    '영어': np.random.randint(60, 101, 10),
    '과학': np.random.randint(60, 101, 10)
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data1)
```

```python
#concat 연습
'''
pandas의 concat은 “여러 개의 Series/DataFrame을 위아래(행) 또는 좌우(열)로 그대로 이어 붙이는 함수”라고 생각하면 됩니다.​

기본 개념
형식: pd.concat(objs, axis=0, ...) 처럼 여러 개의 객체(리스트/튜플) 를 받아서 하나로 합칩니다.​

axis=0이면 행 방향(세로) 으로 위아래로 쌓고, axis=1이면 열 방향(가로) 으로 옆으로 붙입니다.​
ignore_index

기본값 False: 원래 각 DataFrame의 인덱스를 그대로 유지.​

True: 기존 인덱스를 무시하고 0,1,2,...로 새로 다시 매깁니다.​

여러 개를 세로로 붙일 때 인덱스가 겹쳐서 헷갈릴 것 같으면 ignore_index=True를 많이 씁니다.​
'''
```
