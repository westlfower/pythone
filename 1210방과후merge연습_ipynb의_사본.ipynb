```python
import pandas as pd
import numpy as np

np.random.seed(0)

names = [f'학생{i}' for i in range(1, 21)]
student_id = [f'S{i:03d}' for i in range(1, 21)]

# 1학년 명렬표
classes_1 = np.random.choice(['1-1', '1-2'], size=20)
grade1 = pd.DataFrame({
    '학생ID': student_id,
    '이름': names,
    '1학년반': classes_1
})

# 2학년 승급 후 명렬표
classes_2 = np.random.choice(['2-1', '2-2'], size=20)
grade2 = pd.DataFrame({
    '학생ID': student_id,
    '이름': names,
    '2학년반': classes_2
})
```

```python
merged = pd.merge(
    grade1,
    grade2,
    on=['학생ID', '이름'],   # 공통 키
    how='inner'            # 교집합
)

print(merged)
```

```python
merged_left = pd.merge(
    grade1,
    grade2[['학생ID', '2학년반']],  # 필요한 열만
    on='학생ID',
    how='left'
)

print(merged_left)
```
