```python
import pandas as pd
import numpy as np

np.random.seed(42)  # 재현 가능하게

# 반, 성별 랜덤 생성
classes = np.random.choice(['1반', '2반'], size=20)
genders = np.random.choice(['남', '여'], size=20)

# 점수 랜덤 생성 (40~100점)
scores = {
    '국어': np.random.randint(40, 101, size=20),
    '수학': np.random.randint(40, 101, size=20),
    '과학': np.random.randint(40, 101, size=20),
    '영어': np.random.randint(40, 101, size=20)
}

# 이름 생성
names = [f'학생{i}' for i in range(1, 21)]

# 데이터프레임 생성
df = pd.DataFrame({
    '이름': names,
    '반': classes,
    '성별': genders,
    '국어': scores['국어'],
    '수학': scores['수학'],
    '과학': scores['과학'],
    '영어': scores['영어']
})

# 행 순서 섞기
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(df)
```

```python
#학생 순으로 정렬하려면  df.sort_values('국어') #ascending=False
df=df.sort_values('이름').reset_index(drop=True)
```

```python
df
```

```python
#크로스집계 틀 잡기
cross_df=pd.DataFrame({'국어':[0,0],'수학':[0,0],'과학':[0,0],'영어':[0,0]},index=['1반','2반'])
```

```python
cross_df
```

```python
#집계하기
for i in range(len(df)):
  반=df.iloc[i,1]
  국어=df.iloc[i,3]
  수학=df.iloc[i,4]
  과학=df.iloc[i,5]
  영어=df.iloc[i,6]
  cross_df.loc[반, '국어']+=국어
  cross_df.loc[반, '수학']+=수학
  cross_df.loc[반, '과학']+=과학
  cross_df.loc[반, '영어']+=영어
```

```python
cross_df
```

```python
#평균으로 변환하기
for i in range(len(cross_df)):
    for j in range(4):
        cross_df.iloc[i,j]=cross_df.iloc[i,j]/len(df)
```
