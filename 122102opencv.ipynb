## Load Santa Image

### Subtask:
`santa.png` 파일을 OpenCV를 사용하여 불러옵니다.


**Reasoning**:
First, I'll import the OpenCV library as it is required to load the image. Then, I'll use `cv2.imread()` to load the 'santa.png' file and store the result in a variable named `santa_image`.



```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from IPython.display import display
import urllib.request
```

```python
santa_image=cv2.imread('santa.png')
```

```python
from cv2.gapi import inRange
def mosaic_image(image,factor=20):
  h, w =image.shape[:2]
  small=cv2.resize(image,(w//factor,h//factor),interpolation=cv2.INTER_LINEAR)
  return cv2.resize(small,(w,h),interpolation=cv2.INTER_NEAREST)
mosaic_full=mosaic_image(santa_image)
cv2_imshow(mosaic_full)
```
