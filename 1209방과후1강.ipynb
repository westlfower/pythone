```python
import pandas as pd

# 1학년 1반, 2반 수학 시험 성적 예시
df = pd.DataFrame({
    '학번': [1, 2, 3, 4, 5, 6],
    '이름': ['철수', '영희', '민수', '지영', '준호', '수진'],
    '반':   ['1반', '1반', '1반', '2반', '2반', '2반'],
    '수학': [95,   80,   88,   70,   92,   85],
    '성별': ['M', 'F', 'M', 'F', 'M', 'F']
})
print(df)
```

```python
#loc로 이름이 철수인 학생의 수학 점수만 출력
q1=df.loc[df['이름']=='철수','수학']
#iloc로 첫번째행의 모든 정보 출력
q2=df.iloc[0,:]
```

```python
#for + iloc로 크로스 집계 연습
#목표: 반x성별별 인원수 표 만들기
#for문으로 직접집계
반_list=df['반'].unique()
성별_list=df['성별'].unique()
#빈표 (딕셔너리) 만들기
cross={반:{성별: 0 for 성별 in 성별_list} for 반 in 반_list}
```

```python
# 반/성별 목록 만들기
반_list = df['반'].unique()     # ['1반', '2반']
성별_list = df['성별'].unique() # ['M', 'F']

# 빈 표(딕셔너리) 만들기
cross = {반: {성별: 0 for 성별 in 성별_list} for 반 in 반_list}

# for + iloc로 행 하나씩 보며 세기
for i in range(len(df)):
    반 = df.iloc[i, 2]    # 2번 열 = '반'
    성별 = df.iloc[i, 4]  # 4번 열 = '성별'
    cross[반][성별] += 1

cross_df = pd.DataFrame(cross).T
print(cross_df)
```
