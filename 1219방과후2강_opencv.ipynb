```python
import numpy as np
import cv2

h,w= 16, 16
canvas=np.zeros((h,w,3),dtype=np.uint8)
```

```python
canvas
```

```python
BLACK=(0,0,0)
WHITE=(255,255,255)
BLUE=(255,0,0)
GREEN=(0,255,0)
RED=(0,0,255)
YELLOW=(0,255,255)
```

```python
canvas[:,:]=(255,255,200)
```

```python
canvas
```

```python
canvas[4:12,4:12]=BLUE #중앙부분 파란색  사각형
canvas[4:8,4:12]=(200,200,255)#얼굴 영역 조금 밝게
```

```python
canvas
```

```python
canvas[5,6]=WHITE
canvas[5,9]=WHITE
canvas[6,6]=BLACK
canvas[6,9]=BLACK
#눈(흰색+검정)
```

```python
canvas
```

```python
canvas[2,7]=YELLOW
canvas[1,7]=YELLOW
canvas[2,6]=YELLOW
canvas[2,8]=YELLOW
canvas[3,7]=YELLOW
```

```python
canvas
```

```python
#픽셀아트처럼 크게 키우기
scale=20 #1칸을 20*20 필셀로
big_h, big_w = h*scale, w*scale
big_img=cv2.resize(canvas,(big_w,big_h),
                   interpolation=cv2.INTER_NEAREST)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(big_img)
```
