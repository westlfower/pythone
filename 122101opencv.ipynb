```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

height, width=500,500
image=np.ones((height,width,3),dtype=np.uint8)
```

```python
image
```

```python
image=image*255
```

```python
image
```

```python
a, b, c =0.005, -1.0, 250
x_values=np.arange(0,width,1)
y_values=a*x_values**2+b*x_values+c
y_values=height-y_values
y_values=np.clip(y_values,0,height)
y_values=y_values.astype(np.int32)
```

```python
points=np.column_stack((x_values,y_values)).reshape((-1,1,2))
```

```python
points
```

```python
cv2.polylines(image,[points],False,(0,0,255),thickness=2)
cv2.line(image,(50,height-50),(width-50,height-50),(0,0,0),thickness=1)
cv2_imshow(image)
```
